```python
import tensorflow as tf

def cost(hypothesis, Y):
    return tf.reduce_mean(tf.square(hypothesis - Y))

x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]

y_data = [152., 185., 180., 196., 142.]

x1 = tf.Variable(0.0)
x2 = tf.Variable(0.0)
x3 = tf.Variable(0.0)
Y = tf.Variable(0.0)

#[1]은 w에 하나의 값을 가진다는 것을 의미
w1 = tf.Variable(tf.random.normal([1]), name='weight1')
w2 = tf.Variable(tf.random.normal([1]), name='weight2')
w3 = tf.Variable(tf.random.normal([1]), name='weight3')
b = tf.Variable(tf.random.normal([1]), name='bias')


optimizer = tf.optimizers.SGD(learning_rate=1e-5)

for step in range(2001):
    with tf.GradientTape() as tape:
        hypothesis = x1_data * w1 + x2_data * w2 + x3_data * w3 + b
        curr_cost = cost(hypothesis, y_data)

    gradients = tape.gradient(curr_cost, [w1, w2, w3, b])
    optimizer.apply_gradients(grads_and_vars=zip(gradients, [w1, w2, w3, b]))

    if step % 10 == 0:
        print(step, "Cost: ", curr_cost.numpy(), "\nPrediction:\n", hypothesis.numpy())

#이 방법은 잘 사용되지 않음, matrix 이용
```

```python
import numpy as np
import tensorflow as tf

def cost(hypothesis, Y):
    return tf.reduce_mean(tf.square(hypothesis - Y))

filename = 'data-01-test-score.csv'
xy = np.loadtxt(filename, delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
batch_size = 10

#한번에 10개씩 가져와서 x_data와 y_data를 dataset에 저장해줌
dataset = tf.data.Dataset.from_tensor_slices((x_data, y_data)).batch(batch_size)

X = tf.keras.Input(shape=(3,), batch_size=None)
Y = tf.keras.Input(shape=(1,), batch_size=None)

W = tf.Variable(tf.random.normal([3, 1]), name='weight')
b = tf.Variable(tf.random.normal([1]), name='bias')

optimizer = tf.keras.optimizers.SGD(learning_rate=1e-5)

for step in range(2001):
    for batch_x, batch_y in dataset:    #x_data와 y_data의 값을 순차적으로 batch_x, batch_y에 넣어줌
        with tf.GradientTape() as tape:
            hypothesis = tf.matmul(batch_x, W) + b 
            curr_cost = cost(hypothesis, batch_y)
        grads = tape.gradient(curr_cost, [W, b])
        optimizer.apply_gradients(grads_and_vars=zip(grads, [W, b]))
        if step % 10 == 0:
            print(step, "Cost: ", curr_cost.numpy(), "\nPrediction:\n", hypothesis.numpy())
```
