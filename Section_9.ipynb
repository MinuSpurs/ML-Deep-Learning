```python
import tensorflow as tf
import numpy as np


x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)


class XORModel(tf.keras.Model):
    def __init__(self):
        super(XORModel, self).__init__()
        self.layer1 = tf.keras.layers.Dense(units=2, activation='sigmoid', input_shape=(2,))
        self.layer2 = tf.keras.layers.Dense(units=1, activation='sigmoid')

    def call(self, inputs):
        layer1_output = self.layer1(inputs)
        return self.layer2(layer1_output)

model = XORModel()

optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(x_data, y_data, epochs=10001, verbose=0)

for step, (loss, acc) in enumerate(zip(history.history['loss'], history.history['accuracy'])):
    if step % 100 == 0:
        print(f'Step: {step} / Loss: {loss} / Acc: {acc}')

hypothesis = model.predict(x_data)
predicted = (hypothesis > 0.5).astype(np.float32)
accuracy = np.mean(predicted == y_data)

print(f"\nHypothesis:\n{hypothesis} \nPredicted:\n{predicted} \nAccuracy:\n{accuracy}")
```

```python
import tensorflow as tf
import numpy as np

np.random.seed(777)

x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)

class XORModel(tf.keras.Model):
    def __init__(self):
        super(XORModel, self).__init__()
        self.layer1 = tf.keras.layers.Dense(units=10, activation='relu', input_shape=(2,))
        self.layer2 = tf.keras.layers.Dense(units=10, activation='relu')
        self.layer3 = tf.keras.layers.Dense(units=10, activation='relu')
        self.layer4 = tf.keras.layers.Dense(units=1, activation='sigmoid')

    def call(self, inputs):
        layer1_output = self.layer1(inputs)
        layer2_output = self.layer2(layer1_output)
        layer3_output = self.layer3(layer2_output)
        return self.layer4(layer3_output)

model = XORModel()

optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(x_data, y_data, epochs=10001, verbose=0)

for step, (loss, acc) in enumerate(zip(history.history['loss'], history.history['accuracy'])):
    if step % 100 == 0:
        print(f'Step: {step} / Loss: {loss} / Acc: {acc}')

hypothesis = model.predict(x_data)
predicted = (hypothesis > 0.5).astype(np.float32)
accuracy = np.mean(predicted == y_data)

print(f"\nHypothesis:\n{hypothesis} \nPredicted:\n{predicted} \nAccuracy:\n{accuracy}")

```

```python
import tensorflow as tf
import numpy as np


x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=2, activation='relu', input_shape=(2,), kernel_initializer='he_normal'))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid', kernel_initializer='he_normal'))

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])

log_dir = "./logs/xor_logs_r0_01"
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

model.fit(x_data, y_data, epochs=10001, verbose=0, callbacks=[tensorboard_callback])

_, accuracy = model.evaluate(x_data, y_data)
print(f"Accuracy: {accuracy}")

hypothesis = model.predict(x_data)
predicted = (hypothesis > 0.5).astype(np.float32)
print(f"\nHypothesis:\n{hypothesis} \nPredicted:\n{predicted}")

#텐서보드로 확인
```
